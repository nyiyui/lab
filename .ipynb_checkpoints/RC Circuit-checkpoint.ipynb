```python
%matplotlib widget
from math import *
import matplotlib.pyplot as plt
import itertools


def simulate(v_in, timestep=0.01, cycles=100, r=1, c=1):
    t = 0
    v_out = 0
    q = 0
    v_c = 0
    
    data = []
    for i in range(cycles):
        t += timestep
        r_c = timestep/c
        v_out = (r*v_c + r_c*v_in(t))/(r + r_c)
        # update
        i_2 = (v_c-v_out)/r_c
        q += -i_2
        v_c += -i_2*r_c
        data.append((t, v_out, v_c, i_2, q))
    return data

timestep = 0.002
cycles = 1000
v_in = lambda t: sin(t*2**t/4*180/pi)
data = simulate(v_in, timestep=timestep, cycles=cycles)
data_v_in = []
for i in range(cycles):
    t = i*timestep
    data_v_in.append((t, v_in(t)))
plt.ion()
plt.plot(*zip(*data_v_in), color='red', label='v_in')
plt.plot(*itertools.islice(zip(*data), 2), color='blue', label='v_out')
# zipped = list(zip(*data))
# plt.plot(zipped[0], zipped[4], color='green')
plt.legend(loc='lower left')
plt.show()
# plt.close()

'''
timestep = 0.01
r = 1
c = 1
cycles = 100

t = 0
v_in = lambda t: sin(t*t/4*180/pi)
# def v_in(t):
#     t_2 = 18
#     return int(t % t_2 < t_2/2)
#     # if t > 0.5:
#     #     return 1
#     # else:
#     #     return 0
v_out = 0

v_c = 0

data = []
v_c_data = []leg = plt.legend(loc='upper center')
for i in range(cycles):
    t += timestep
    r_c = timestep/c
    v_out = (r*v_c + r_c*v_in(t))/(r + r_c)
    v_c_data.append((t, v_c))
    # update
    i_2 = (v_c-v_out)/r_c
    v_c += -i_2*r_c
    data.append((t, v_out))


data_v_in = []
for i in range(cycles):
    t = i*timestep
    data_v_in.append((t, v_in(t)))
plt.plot(*zip(*data_v_in), color='red')
plt.scatter(*zip(*data))
plt.scatter(*zip(*v_c_data))
plt.show()
plt.close()
'''
```

## Chirp

```python
timestep = 0.002
cycles = 1000
v_in = lambda t: sin(t*2**t/4*180/pi)
data = simulate(v_in, timestep=timestep, cycles=cycles)
data_v_in = []
for i in range(cycles):
    t = i*timestep
    data_v_in.append((t, v_in(t)))
plt.close()
plt.clf()
plt.ion()
plt.plot(*zip(*data_v_in), color='red', label='v_in')
plt.plot(*itertools.islice(zip(*data), 2), color='blue', label='v_out')
plt.legend(loc='lower left')
plt.show()
```

## $\sin x$

## Carrier

See how the modulated signal contains two frequencies:

```python
import numpy as np
from matplotlib import gridspec


x = np.arange(0,2*np.pi,0.01)   # start,stop,step
s = np.sin(x)
c = np.sin(10*x)
m = s*c
plt.ion()
fig = plt.figure("Carrier Signal")
gs = gridspec.GridSpec(4, 1, height_ratios=[1]*4)

sp0 = plt.subplot(gs[0])
sp0.plot(x, s, label='signal')
sp0.legend(loc='lower right')

sp1 = plt.subplot(gs[1])
sp1.plot(x, c, label='carrier')
sp1.legend(loc='lower right')

sp2 = plt.subplot(gs[2])
sp2.plot(x, m, label='modulated')
sp2.legend(loc='lower right')

'''
timestep = 0.002
cycles = 1000
v_in = lambda x: sin(x)*sin(10*x)
data = simulate(v_in, timestep=timestep, cycles=cycles, r=1e-18, c=1e9)
data_v_in = []
for i in range(cycles):
    t = i*timestep
    data_v_in.append((t, v_in(t)))
sp3 = plt.subplot(gs[3])
# sp3.plot(*zip(*data_v_in), color='red', label='v_in')
sp3.plot(*itertools.islice(zip(*data), 2), color='blue', label='low-pass')
sp3.legend(loc='lower right')
'''

plt.subplots_adjust(hspace=.0)
plt.show()
```
