## Case 1a - Free Fall

```python
import numpy as np
import matplotlib.pyplot as plt
import math


# Define mass if you need it
m = 0.1

# Define here the Force Function
def free_fall(t, x, v):
  '''
  This defines the force on the object.
  This force function models a free fall, so only gravitational force is applied.
  '''
  # gravity is defined as (g > 0) -g N/kg
  return -9.806*m

F = free_fall

# Choose a time interval
t1 = 0
t2 = 10

# Choose the number of steps of integration
n = 100

# Find the time step dt
dt = (t2 - t1)/n

# make arrays for t, x, v, and a
t = np.arange(t1, t2 + dt, dt) # Time Array (n + 1 elements)
x = np.zeros(len(t)) # Position Array
v = np.zeros(len(t)) # Velocity Array
a = np.zeros(len(t)) # Acceleration Array

# Define the initial conditions x(0) and v(0)
x[0] = 1
v[0] = 0

# Explicit Euler's Method

for i in range(n): # from i = 0 to i = n - 1
    a[i] = F(t[i], x[i], v[i])/m
    v[i + 1] = v[i] + a[i] * dt
    x[i + 1] = x[i] + (v[i] + v[i + 1])*dt/2
    
a[n] = a[n-1] # to fill the last element of the a array

plt.figure(figsize = (12, 8))
plt.plot(t, x, '-', label = 'Position')
plt.plot(t, v, '-', label = 'Velocity')
plt.plot(t, a, '-', label = 'Acceleration')

plt.title('Position, Velocity, and Acceleration versus Time')
plt.xlabel('t')
plt.ylabel('x, v, a')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

## Case 4: Gravity and air resistance with a floor

```python
import numpy as np
import matplotlib.pyplot as plt
import math


# Define mass if you need it
m = 0.1

# Choose a time interval
t1 = 0
t2 = 10

# Choose the number of steps of integration
n = 100

# Find the time step dt
dt = (t2 - t1)/n

# Define here the Force Function
def sum_force(*forces):
  "This function returns a force function with the sum of the given forces."
  return lambda t, x, v: sum(force(t, x, v) for force in forces)

def free_fall(t, x, v):
  '''
  This defines the force on the object.
  This force function models a free fall, so only gravitational force is applied.
  '''
  # gravity is defined as (g > 0) -g N/kg
  return -9.806*m

def floor(force, /, floor):
  "This simulates an elastic floor and object."
  prev = None
  def result(t, x, v):
    nonlocal prev
    if x > floor:
      prev = (t, x, v)
      return force(t, x, v)
    elif x <= floor and prev and prev[1] > floor:
      assert v < 0
      prev = (t, x, v)
      return (2*abs(v))/dt*m
    else:
      prev = (t, x, v)
      return -abs(v)/dt*m if v < 0 else force(t, x, v)
  return result

def air_resistance(t, x, v):
  '''
  This force function models a proportional (to velocity) air resistance.
  '''
  b = 0.05
  return -b*v

def sinusoidal(t, x, v):
  '''
  This force function models a sinusoidal force.
  '''
  a = 1
  b = 1
  return a*math.sin(b*t)

F = floor(sum_force(air_resistance, free_fall), floor=-50)

# make arrays for t, x, v, and a
t = np.arange(t1, t2 + dt, dt) # Time Array (n + 1 elements)
x = np.zeros(len(t)) # Position Array
v = np.zeros(len(t)) # Velocity Array
a = np.zeros(len(t)) # Acceleration Array

# Define the initial conditions x(0) and v(0)
x[0] = 1
v[0] = 0

# Explicit Euler's Method

for i in range(n): # from i = 0 to i = n - 1
    a[i] = F(t[i], x[i], v[i])/m
    v[i + 1] = v[i] + a[i] * dt
    x[i + 1] = x[i] + (v[i] + v[i + 1])*dt/2
    
a[n] = a[n-1] # to fill the last element of the a array

plt.figure(figsize = (12, 8))
plt.plot(t, x, '-', label = 'Position')
plt.plot(t, v, '-', label = 'Velocity')
plt.plot(t, a, '-', label = 'Acceleration')

plt.title('Position, Velocity, and Acceleration versus Time')
plt.xlabel('t')
plt.ylabel('x, v, a')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
